```python
import pandas as pd
from datasets import load_dataset
import yaml
import os
import io

# Load Hugging Face text dataset
ds = load_dataset("melisekm/natural-disasters-from-social-media")
# Filter the text dataset for 'floods'
flood_ds = ds.filter(lambda example: 'floods' in example['event_type_detail'])

# Create a clean DataFrame from the 'text' and 'event_type_detail' columns
# This is where we ensure the columns exist before processing.
df_text = pd.DataFrame({
    'text': flood_ds['train']['text'],
    'disaster_type': flood_ds['train']['event_type_detail']
})

# Create urgency labels for the text data
def label_urgency(text):
    urgent_keywords = ['help', 'stuck', 'stranded', 'rescue', 'emergency', 'trapped']
    if any(keyword in text.lower() for keyword in urgent_keywords):
        return 1  # Urgent
    else:
        return 0  # Informational

df_text['urgency_label'] = df_text['text'].apply(label_urgency)

print("DataFrame with urgency labels created successfully.")
print(df_text.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming 'df' is your pandas DataFrame from the Hugging Face dataset
# from Step 3 of the previous notebook.

# 1. Data Transformation: Convert text to numerical features using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
X = vectorizer.fit_transform(df['text']) # Corrected: 'text' instead of 'clean_text'

# 2. Feature Selection: Use 'target' (binary) as labels
y = df['target']

# 3. Select top features
selector = SelectKBest(chi2, k=1000)
X_selected = selector.fit_transform(X, y)

print(f"Original feature matrix shape: {X.shape}")
print(f"Selected feature matrix shape: {X_selected.shape}")

# 4. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# 5. Train a simple model (optional, but good for demonstrating the process)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# 6. Evaluate the model
y_pred = model.predict(X_test)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```
